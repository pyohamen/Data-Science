# Loan 데이터 분류나무

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('loan.csv')
df
```

```python
df.isnull().sum().sum()
```

```python
df.groupby('job').mean()
```

```python
# self sorting..
df['job'] = df['job'].replace({'Office':0, 'ProfExe':1, 'Other':2, 'Mgr':3, 'Self':4, 'Sales':5})
df.groupby('job').mean()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cmap='coolwarm',annot=True)
plt.show()
```

```python
plt.figure(figsize=(15,15))
for i in range(0,9):
    plt.subplot(3,3,i+1)
    sns.kdeplot(data=df, x=df.columns[1:][i], hue="y")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='DTI', y='delinquency', hue='y')
plt.show()
```

### 나무 모형

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
X = df.drop(['y'], axis=1)
y = df['y']
```

### Default 트리

```python
c_tree = DecisionTreeClassifier(random_state=0) 
c_tree.fit(X, y)
plt.figure(figsize=(20,15))
plot_tree(c_tree, filled=True) 
plt.show()
```
