# Basic regression: Predict fuel efficiency 

https://www.tensorflow.org/tutorials/keras/regression

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


# Make numpy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
```

```python
#  download and import the dataset using pandas:
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'Model Year', 'Origin']

raw_dataset = pd.read_csv(url, names=column_names,
                          na_values='?', comment='\t',
                          sep=' ', skipinitialspace=True)
```

```python
dataset = raw_dataset.copy()
dataset.tail()
```

```python
# clean data set 
# drop missing values
dataset.isna().sum()
```

```python
dataset = dataset.dropna()
```

```python
# 더미 변수화
dataset['Origin'] = dataset['Origin'].map({1: 'USA', 2: 'Europe', 3: 'Japan'})
dataset = pd.get_dummies(dataset, prefix='', prefix_sep='')
dataset.tail()
```

```python
# split train and test data
train_dataset = dataset.sample(frac=0.8, random_state=0) # 80% 를
test_dataset = dataset.drop(train_dataset.index)
```

```python
# Separate the target value, the "label", from the features. This label is the value that you will train the model to predict.

# input 으로 사용할 X 변수
train_features = train_dataset.copy()
test_features = test_dataset.copy()

# target 으로 보는 것만을
train_labels = train_features.pop('MPG')
test_labels = test_features.pop('MPG')
```

```python
# normalization ( 정확도를 높이기 위한 정규화 )
# 입력층에서 진행
normalizer = preprocessing.Normalization()
normalizer.adapt(np.array(train_features))
print(normalizer.mean.numpy())
```

```python
# default 가 linear
model = tf.keras.models.Sequential([
      normalizer,
      layers.Dense(64,  activation='relu'),
      layers.Dense(64, activation='relu'), # 은닉층 64개
      layers.Dense(1) # 출력층 1개
])

model.summary()
```

```python
# output 이 수치형 자료이기 때문에 mean_absolute_error 씀, binary_crossentropy 안됨
model.compile(loss='mean_absolute_error', optimizer=tf.keras.optimizers.Adam(0.001))
```

```python
# fit the keras model on the dataset
model.fit(train_features, train_labels, epochs=100)
```

```python
# 모형 적합에 사용되지 않은 test data 를 예측
test_results = model.evaluate(test_features, test_labels, verbose=0)
test_results
```

```python
test_predictions = model.predict(test_features).flatten()

a = plt.axes(aspect='equal')
plt.scatter(test_labels, test_predictions)
plt.xlabel('True Values [MPG]')
plt.ylabel('Predictions [MPG]')
lims = [0, 50]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
```
